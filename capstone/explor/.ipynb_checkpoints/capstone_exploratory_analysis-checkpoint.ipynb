## Load full dataset

The dataset is roughly 40K rows, each row representing a slice-radiologist combination.

```python
import pandas as pd
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(color_codes=True)
```

```python
data = pd.read_csv("all_slices.csv")
```

```python
data_semantic = data[['subtlety', 'internalStructure', 'calcification',
                         'sphericity', 'margin', 'lobulation', 
                         'spiculation', 'texture', 'imageSOP_UID','malignancy']]
data_semantic.head()
```

#### create dataframes using the min, max and mean of the semantic values for each slice

```python
data_max = data_semantic.groupby("imageSOP_UID").max()
data_min = data_semantic.groupby("imageSOP_UID").min()
data_mean = data_semantic.groupby("imageSOP_UID").mean()
print "Unique Slices:", len(np.unique(data_semantic["imageSOP_UID"]))
```

## Semantic Attribute Analysis

Below we begin to look at the variance among the semantic attributes, by using a dictionary called tmp to capture the vector of scores for each slice and then storing the variance for the semantic ratings for each slice in list m.  Next we look at the distribution of variance, as well as the distribution using the min, max and mean for each slice.  Note, some of the slices had 1 rating for semantic values, so there is no variance.  When filtering these, the results for similar so it was left out for brevity.

### Malignancy

#### variance

```python
tmp = {}
m = []

for i,r in data_semantic.iterrows(): tmp[r['imageSOP_UID']] = []
for i,r in data_semantic.iterrows(): tmp[r['imageSOP_UID']].append(int(r['malignancy']))
for k,v in tmp.items(): m.append(np.var(v))

sns.distplot(m, kde=False);
```

#### histogram using the max

```python
sns.distplot(data_max["malignancy"], kde=False);
```

#### histogram using the min

```python
sns.distplot(data_min["malignancy"], kde=False);
```

#### histogram using the mean

```python
sns.distplot(data_mean["malignancy"], kde=False);
```
